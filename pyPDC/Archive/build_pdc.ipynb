```python
import pdc
pdc1 = pdc.Pdc(pdc_id=7, pmu_ip="127.0.0.1", pmu_port=4712)
```

```python


# pdc = Pdc(pdc_id=7, pmu_ip="127.0.0.1", pmu_port=4712)

pdc1.run()  # Connect to PMU
header = pdc1.get_header()  # Get header message from PMU
config,rawdata = pdc1.get_config()  # Get configuration from PMU
cfgpkt = synphasor(rawdata)
cfgpkt.show()
set_cfg_data(cfgpkt)

pdc1.start()  # Request to start sending measurements

while True:
    obj,raw = pdc1.get()  # Keep receiving data
    print("Packet type: ",type(raw))
    pkt = synphasor(raw)
    pkt.show()
    if not obj:
        pdc1.quit()  # Close connection
        break
```
